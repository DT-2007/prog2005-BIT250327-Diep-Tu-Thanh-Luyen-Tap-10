```python
strings = []
for i in range(5):
    string_input = input(f"Nhập chuỗi ký tự thứ {i+1}: ")
    strings.append(string_input)

n = len(strings)
for i in range(n):
    for j in range(0, n-i-1):
        if len(strings[j]) < len(strings[j+1]):
            strings[j], strings[j+1] = strings[j+1], strings[j]
            print("Danh sách hiện tại:", strings)

print("Danh sách sau khi sắp xếp theo độ dài giảm dần:", strings)
```
