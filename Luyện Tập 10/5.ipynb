```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.figure(figsize=(10,5))

# subplot 1
plt.subplot(1,2,1)
plt.plot(x, x**2)
plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")

# subplot 2
plt.subplot(1,2,2)
plt.plot(x, np.sqrt(x))
plt.title("y = sqrt(x)")
plt.xlabel("x")
plt.ylabel("y")

plt.tight_layout()
plt.show()
```
