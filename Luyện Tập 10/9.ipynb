```python
class Nguoi:
    dem = 0

    def __init__(self, ten, tuoi):
        if tuoi < 0:
            raise ValueError("Tuổi không hợp lệ")
        self._ten = ten
        self._tuoi = tuoi
        Nguoi.dem += 1

    # getter
    def get_tuoi(self):
        return self._tuoi

    # setter
    def set_tuoi(self, tuoi):
        if tuoi < 0:
            raise ValueError("Tuổi phải >=0")
        self._tuoi = tuoi

    def __str__(self):
        return f"Tên: {self._ten}, Tuổi: {self._tuoi}"

    def xin_chao(self):
        print("Xin chào!")

    @classmethod
    def so_luong(cls):
        return cls.dem

    @staticmethod
    def thong_tin():
        return "Đây là lớp Nguoi"

    def __eq__(self, other):
        return self._tuoi == other._tuoi


class SinhVien(Nguoi):
    def __init__(self, ten, tuoi, diem):
        super().__init__(ten, tuoi)
        if diem < 0 or diem > 10:
            raise ValueError("Điểm không hợp lệ")
        self.diem = diem

    def __str__(self):
        return super().__str__() + f", Điểm: {self.diem}"


# Test
sv1 = SinhVien("A", 20, 8)
sv2 = SinhVien("B", 20, 9)

print(sv1)
sv1.xin_chao()
print(Nguoi.so_luong())
print(Nguoi.thong_tin())
print("So sánh:", sv1 == sv2)
```
